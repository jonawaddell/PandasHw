```python
#import 
import pandas as pd
import numpy as np
```

```python
#import json file
raw_json = "Documents/GitHub/RUTSOM201801DATA5-Class-Repository-DATA/04-Pandas/2-homework/Instructions/HeroesOfPymoli/purchase_data.json"
```

```python
#read json file
json = pd.read_json(raw_json)
#read json file as DataFrame
df_json = pd.DataFrame(json)
print(json.columns)
```

```python
#format all floats to currency
pd.options.display.float_format = '${:,.2f}'.format
```

```python
#Player Count
player_count = df_json["SN"].value_counts().count() 

count_summary= pd.DataFrame({'Total Number of Players':[player_count]})
count_summary
```

```python
#Purchasing Analysis (Total)
#Number of Unique Items
unqitems_df_json = df_json['Item ID'].value_counts().count()

#Average Purchase Price
avg_df_json = df_json['Price'].mean()

#Total Number of Purchases
totpur_df_json = df_json['SN'].count()

#Total Revenue
totrev_df_json = df_json['Price'].sum()
#-----------------------
Purchasing_Analysis_Sum = pd.DataFrame({'Number of Unique Items':[unqitems_df_json],
                                       'Average Purchase Price':[avg_df_json],
                                       'Total Number of Purchases':[totpur_df_json],
                                       'Total Revenue':[totrev_df_json],})
Purchasing_Analysis_Sum
```

```python
#Gender Demographics values
gender_count = df_json['Gender'].value_counts()
gender_percent = df_json['Gender'].value_counts()/780
#dataframe for gender demographic 
gender_demo_sum = pd.DataFrame({'Percent of Players':gender_percent * 100,
                              'Total Count': gender_count})
gender_demo_sum = gender_demo_sum.round(2)
gender_demo_sum
```

```python
#Purchase Analysis(Gender)
#get values
gender_count = df_json.groupby(['Gender']).count()['Price'].rename("Purchase Count")
gender_average = df_json.groupby(['Gender']).mean()['Price'].rename("Average Purchase Value")
gender_purchase_total = df_json.groupby(['Gender']).sum()['Price'].rename("Total Purchase Value")
#find normalized total
normalized = gender_purchase_total / gender_demo_sum['Total Count']
#create data frame
gender_data = pd.DataFrame({"Normalized Total": normalized, 
                            "Purchase Count": gender_count, 
                            "Total Purchase Value": gender_purchase_total, 
                            "Average Purchase Value": gender_average})
gender_data        
```

```python
#Age Demographics
#bin ages
bins = [0,10,14,19,24,29,34,39,50]
ages_bin = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
df_json["Age Demographics"] = pd.cut(df_json['Age'], bins, labels=ages_bin)
df_json["Age Demographics"].head()
```

```python
age_count = df_json['Age Demographics'].value_counts()
age_percent = df_json['Age Demographics'].value_counts()/780

age_demo_sum = pd.DataFrame({'Percent of Players':age_percent * 100,
                            'Total Count': age_count})
age_demo_sum
```

```python
age_pur_count = df_json.groupby(["Age Demographics"]).count()['Price']
age_avg_price = df_json.groupby(["Age Demographics"]).mean()['Price']
age_total_value = df_json.groupby(["Age Demographics"]).sum()['Price']

age_normalized = age_total_value/age_demo_sum['Total Count']

age_data = pd.DataFrame({"Normalized Total": age_normalized, 
                            "Purchase Count": age_pur_count, 
                            "Total Purchase Value": age_total_value, 
                            "Average Purchase Value": age_avg_price})
age_data
```

```python
#top spender
#get values
user_total = df_json.groupby(["SN"]).sum()["Price"].rename("Total Purchase Amount")
user_average = df_json.groupby(["SN"]).mean()["Price"].rename("Average Purchase Price")
user_count = df_json.groupby(["SN"]).count()["Price"].rename("Purchase Count")
#create dataframe
user_data = pd.DataFrame({"Total Purchase Amount": user_total,
                          "Average Purchase Price": user_average,
                          "Purchase Count": user_count})

# Convert to Data Frame
user_data.sort_values("Total Purchase Amount", ascending=False).head(5)
```

```python
#Most Popular Item
#get values
popitem_total = df_json.groupby(['Item ID', 'Item Name']).count()["Price"]
popitem_price = df_json.groupby(['Item ID', 'Item Name']).mean()['Price']
popitem_purchase_value = df_json.groupby(['Item ID', 'Item Name']).sum()['Price']
#create dataframe
popitem_data = pd.DataFrame ({'Purchase Count': popitem_total,
                             'Item Price': popitem_price,
                             'Total Purchase Value': popitem_purchase_value})

popitem_data.sort_values('Purchase Count', ascending=False).head(5)
```

```python
#Most Profitable  Item
#get values
popitem_total = df_json.groupby(['Item ID', 'Item Name']).count()["Price"]
popitem_price = df_json.groupby(['Item ID', 'Item Name']).mean()['Price']
popitem_purchase_value = df_json.groupby(['Item ID', 'Item Name']).sum()['Price']
#create dataframe
popitem_data = pd.DataFrame ({'Purchase Count': popitem_total,
                             'Item Price': popitem_price,
                             'Total Purchase Value': popitem_purchase_value})

popitem_data.sort_values('Total Purchase Value', ascending=False).head(5)
```
